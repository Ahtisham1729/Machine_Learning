```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mglearn
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

# Datasets

1. Forge dataset : for classification type algorithms

```python

# generate dataset
X, y = mglearn.datasets.make_forge()
# plot dataset
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(["class 0", "class 1"], loc=4)
plt.xlabel("First feature")
plt.ylabel("Second feature")
print(f"x.shape: {X.shape}")
```

2. Wave dataset : For Regression Algorithms

```python
X,y = mglearn.datasets.make_wave(n_samples=40)
plt.plot(X, y, 'o')
plt.xlabel("Feature")
plt.ylabel("Target")
```

3. Breast Cancer dataset

```python
from sklearn.datasets import load_breast_cancer
# load dataset
cancer = load_breast_cancer()
# print dataset description
#print(cancer.DESCR)
# print feature names
#print(cancer.feature_names)
# print target names
#print(cancer.target_names)
# print cancer key
print(f"cancer.keys(): \n{cancer.keys()}")
print(f"Shape of cancer data: {cancer.data.shape}")
counts = {
    str(n):  # cast to Python str
    int(v)   # cast to Python int
    for n, v in zip(cancer.target_names, np.bincount(cancer.target))
}
print(f"Sample counts per class:\n{counts}")
print(f"feature names:\n{cancer.feature_names}")
```
